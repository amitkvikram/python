```python
# PLOT sin cos x^2 sin^2x
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
# sns.set()

fig = plt.figure(figsize=(10,10))
arr= np.arange(-1, 10, 0.1)

ax1 = fig.add_subplot(221,xlabel = "x", title = 'sin')
ax1.set_xlabel('x',fontsize = 20)
ax1.scatter(arr, np.sin(arr),label= "amit", s = 80)   #s stands for marker size
ax1.legend()

ax2 = fig.add_subplot(222, xlabel = 'x', title = 'cos')
ax2.plot(arr, np.cos(arr),'r^',label= "amit",markersize = 2)
ax2.set_xlabel('x',fontsize = 20)
ax2.legend()

ax3 = fig.add_subplot(223, xlabel = 'x',title = 'x^2')
ax3.plot(arr,np.square(arr), label = "square")
ax3.legend()

ax4 = fig.add_subplot(224, xlabel = 'x', title = 'sin^2')
ax4.plot(arr, np.square(np.sin(arr)), label = "sin^2")
# ax4.axis([0, 20, 0, 5])
ax4.legend()
# ax4.*?
ax4.grid(True, which = "both")
plt.tight_layout()
plt.xticks(rotation = 90)
ax4.xaxis.set_ticks(np.arange(0.5,1000000,10))
ax4.set_yticks(np.arange(0,1.5,0.5))
plt.show()
```
