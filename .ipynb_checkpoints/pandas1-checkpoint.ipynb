### Series

```python
import pandas as pd
import numpy as np
# web_stats = {'Day':[1,2,3,4,5,6],
#              'Visitors':[43,34,65,56,29,76],
#              'Bounce Rate':[65,67,78,65,45,52]}
s1 = pd.Series(np.random.randn(5), index = ['a','b','c', 'd','e'])
print(s1)
print(s1.index)
display(type(s1))
```

```python
d = {'a':11, 'b':12,'c':13, 'd':14}
s2 = pd.Series(d)
print(s2)
```

```python
print(s1>s1.median())
print(s1[s1>s1.median()])
```

```python
print(s1[:3])
print(s1[['a', 'b', 'c']])
print(s1[[0,1,2]])
```

```python
import matplotlib.pyplot as plt

s1.plot()
plt.show()
```

### Dataframe

```python
web_stats = {'Day':[1,2,3,4,5,6],
             'Visitors':[43,34,65,56,29,76],
             'Bounce Rate':[65,67,78,65,45,52]}
df1 = pd.DataFrame(web_stats, index=['a','b','c','d','e','f'])
display(df1)
display(df1.index)
print(df1.columns)
display(df1[['Day','Visitors']])
display(type(df1['Day']))
display(df1['Day']['a'])
```

### Column Selection, Deletion, Addition

```python
df = pd.read_csv('test.csv')
display(df.head())
```

```python
print(df.index)
print(df.columns)
display(df['Fare'].head())
```

```python
df['temp1'] = df['Age']*df['SibSp']
display(df.head())
```

```python
del df['temp1']
```

```python
display(df.head())
```

```python
temp1 = df.pop('Cabin')
display(df.head())
display(temp1.head())
```

```python
df.insert(9,'Cabin',temp1)
display(df.head())
```

```python
df.plot()
plt.show()
```

```python
df.plot(kind = "scatter", x='SibSp', y = 'Age')
plt.show()
```

```python
df_temp = df.assign(temp = lambda x:x['Age']/x['Fare'])
display(df_temp.head())
display(df.head())
```
