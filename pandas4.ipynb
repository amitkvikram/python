```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame({'a': [0,1,2, 8], 'b':[None,'B1','None','B3'], 'c': ['C0', np.nan,'C2', 'C3'], 'd':
                   ['D0','D1','D2','D3']},index = ['i1','i2','i3','i4'])

df2 = pd.DataFrame({'a': [1,4,6,2], 'b':['B4','B5','B5','B5'], 'c': ['C4', 'C5','C6', 'C7'], 'd':
                   ['D4','D5','D6','D7']},index = ['i1','i2','i3','i4'])

df3 = pd.DataFrame({'a': ['A8','A9','A10','A11'], 'b':['B8','B9','B10','B11'], 'c': ['C8', 'C9','C10', 'C11'], 
                    'd':['D8','D9','D10','D11']})
display(df1)
```
